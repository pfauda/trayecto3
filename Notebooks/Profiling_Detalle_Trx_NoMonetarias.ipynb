```python
import pandas as pd
import pandas_profiling
import os

DIR = os.path.abspath(r'..\Data\Export') + '\\'
```

```python
uc = [
    'TIPO_CANAL',
    'CANAL',
    'GRUPO_OPERACION',
    'TIPO_OPERACION',
    'TIPO_OPERACION2',
    'ORIGEN',
    'CANAL_RATIO',
    'CLASIF_TRX'
]

tc = dict(zip(uc, [str] * len(uc)))
```

```python
file = DIR + 'Detalle_Trx_NoMonetarias_202204.csv'
df = pd.read_csv(file, index_col=False, usecols=uc, dtype=tc).sample(200_000) #Tomo una muestra sino pandas_profiling explota
```

```python
# Analizamos
df.reset_index(drop=True, inplace=True)
report = pandas_profiling.ProfileReport(df)
report
```
