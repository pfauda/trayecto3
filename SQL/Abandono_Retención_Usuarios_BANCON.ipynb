## Cruce de tablas de Usuarios de platadorma y Crosseling

El propósito de esta operación es obtener una sola tabla resumida con todos los datos de ambas tablas y que podamos calcular con facilidad la columna de dias de no uso de la plataforma.

```
df_estudio = pd.merge(df_Usuarios_PD, df_Prod_Usrs_PD, how='inner', on = 'ID_USER')
```

```
# Borro las columnas que considero a priori que no son relevantes
delCols = ["RELACION", "TIPOIMPN", "NATUJURI", "SUCURSAL", "BCRASECT", "BCRASEGO", "CROSS_EMPR", "CROSS_IND"]
df_estudio.drop(delCols, axis=1, inplace=True)
```

```
# Calculo el saldo total como la suma de todas las columnas de saldo (**Revisar luego como utilizamos este dato**)
df_estudio["SALDO_TOTAL"] = df_estudio["SALDO_COMPUTO_SIST_2"] + df_estudio["SALDO_COMPUTO_SIST_CC_MB"] + df_estudio["SALDO_COMPUTO_SIST_CA_P"] + df_estudio["SALDO_COMPUTO_SIST_CC_P"] + df_estudio["SALDO_COMPUTO_SIST_FCI"]

delCols = ["SALDO_COMPUTO_SIST_2", "SALDO_COMPUTO_SIST_CC_MB", "SALDO_COMPUTO_SIST_CA_P", "SALDO_COMPUTO_SIST_CC_P", "SALDO_COMPUTO_SIST_FCI"]
df_estudio.drop(delCols, axis=1, inplace=True)
```

```
df_estudio.head()
```

```
# Por el momento solo nos quedamos con el último mas (Abril)
df_estudio = df_estudio.loc[df_estudio['PERIODO'] == '202204']
```

```
# 
df_estudio["FUM"] = df_estudio["FUM"].apply(pd.to_datetime)
df_estudio[["FECHA_CREACION", "LAST_LOGIN", "FUM"]].info()
print('\n')
df_estudio[["FECHA_CREACION", "LAST_LOGIN", "FUM"]].head()
```

```
# Calculo la cantidad de dias que transaccionó en el banco desde su ultimo acceso a la plataforma 
df_estudio["DIAS_SIN_TRX_PD"] = (df_estudio["FUM"] - df_estudio["LAST_LOGIN"]).astype('timedelta64[D]')
df_estudio["DIAS_SIN_TRX_PD"] = np.where(df_estudio["DIAS_SIN_TRX_PD"] > 0, df_estudio["DIAS_SIN_TRX_PD"], 0)

# Calculo Antiguedad en dias la plataforma 
df_estudio["ANTIGUEDAD_PD"] = (df_estudio["LAST_LOGIN"] - df_estudio["FECHA_CREACION"]).astype('timedelta64[D]')
df_estudio[["FECHA_CREACION", "LAST_LOGIN", "FUM", "ANTIGUEDAD_PD", "DIAS_SIN_TRX_PD"]].head()
```

```
# Filtro los registros sin antiguedad en la plataforma
df_estudio = df_estudio.loc[df_estudio['ANTIGUEDAD_PD'] != 0]
df_estudio[["FECHA_CREACION", "LAST_LOGIN", "FUM", "ANTIGUEDAD_PD", "DIAS_SIN_TRX_PD"]].head()
```

```
# Filtro los registros que no registran dias de no uso de la PD
df_estudio = df_estudio.loc[df_estudio['DIAS_SIN_TRX_PD'] != 0]
df_estudio.shape
```

```
df_estudio[["FECHA_CREACION", "LAST_LOGIN", "FUM", "ANTIGUEDAD_PD", "DIAS_SIN_TRX_PD"]].sample(5)
```

```
# Asumo que abandonó la PD si lleva mas de 60? dias sin ingresar a la plataforma
#  digital desde el último movimiento registrado
#df_estudio["ABANDONO_PD"] = np.where(df_estudio["DIAS_SIN_TRX_PD"] > 90, 'Sí', 'No') -> me da warning
df_estudio["ABANDONO_PD"] = ['Sí' if _dias > 90 else 'No' for _dias in df_estudio["DIAS_SIN_TRX_PD"]]
```

```
df_estudio[["FECHA_CREACION", "LAST_LOGIN", "FUM", "ANTIGUEDAD_PD", "DIAS_SIN_TRX_PD","ABANDONO_PD"]].sample(5)
```

```
plot = df_estudio["ABANDONO_PD"].value_counts().plot(kind='bar', title="Abandono Plataforma")
```
